```python
#When we have inferSchema = True ,data type is infered as its type of data present in that column else it is consider as string.
df = spark.read.csv("/Volumes/pyspark_learning/learn_by_doing_db/volume_files/order/file.csv",header=True,inferSchema=True)

display(df)
```

```python
df.printSchema()
```

```python
df = spark.read.format("csv")\
    .option("header","true")\
    .option("inferSchema","true")\
    .load("/Volumes/pyspark_learning/learn_by_doing_db/volume_files/order/file.csv")
display(df)
#To print schema of data
df.printSchema()
```
