### What is databrick ?
* Databricks is **unified analytics and data engineer platform** that lets us process ,and analyze all kind of data in one place.
* Unified Analytics Platform = data is managed, process and analyze on same platform
* It's built on top of Apache spark but add many enterprise features, include:-
    * Collaborative workspaces for data engineers, data scientists ,and analysts.
    * Support for data lakes, data warehouses, and machine learning.
    * Delta Lake technology for reliability ,governance, and speed.

##### What is Apache Spark ?
* Apache spark is :-
    * An open-source distributed computing engine for processing large datasets.
    * Very fast compared to older frameworks like Hadoop MapReduce.
    * Can run batch, straeming, SQL, ML and graph workloads.
    * Requires you to manage clusters, storage, integrations, and scaling manually.

##### Why Databricks is Different from just using Spark ?
* Spark alone = Fast distributed procesing, but we must manage clusters, storage, and integrayions manually.
* Databricks = adds a friendly UI, auto-cluster management, collaborative notebooks, optimized connectors, governance tools, ML lifecycle support ,and Delta Lake for reliable data.
* It's like the difference b/w owning a powerful engine vs a fully built sports car - same power, but one is ready to drive.

1. Setup & Infrastructure :-
  * Spark Alone -> We install, configure, and manage our own clusters(on-premise or cloud).
  * Databricks -> Fully managed -- just click to craete clusters, no manual setup.

2. Cluster Management :
  * Spark Alone -> We must manually scale up/down ,monitor, and shut down clusters.
  * Databricks -> Auto-scaling cluster that start/stop based on workload to save cost.

3. Data Storage :
  * Spark Alone -> Read/Write from external storage (S3, HDFS, Azure Blob), but no built-in transactional layer.
  * Databricks -> Uses Delta Lake for ACID transactions ,schema enforcement, and time travel.
4. Data Governance & Security :
  * Spark Alone -> We handle permissions manually at the file system or storage level.
  * Databricks -> Unity Catalog provides centralized data access control, auditing, and compliance.

5. Collaboration :
  * Spark Alone -> No built-in multi-user interface-sharing requires exporting scripts or files.
  * Databricks -> Collaborative notebooks where multiple user can work.

6. Workflow Automation :
  * Spark Alone -> Requires external schedulers(Airflow,cron) for automation.
  * Databricks ->Built-in jobs scheduler with monitoring, retry policies, and alerts.

7. Machine Learning :
  * Spark Alone -> MLlib for basic ML, but no built-in experiment tracking or deploymrnnt tools
  * Databricks -> MLflow integrated for ML lifecycle management (tracking ,registry ,deployment).

8. Optimization 
  * Spark Alone -> We manually tune configurations for performance.
  * Databricks -> Auto-optimizations(e.g Photon engine for faster SQL ,auto-tuning).
9. Integrations 
  * Spark Alone -> We setup JDBC/ODBC drivets manually for BI tools.
  * Databricks -> Native connectors to Power BI, Tableau, Looker, Snowflake ,etc.


### What is Databricks Lakehouse ?
  * The Lakehouse is Databrick's data architecture that combines the best parts of a Data Lake and Data Warehouse in a single system.
  * Data Lake strength -> Store all kinds of data(structured, semi-structured, unstructured) cheaply in cloud storage.
  * Data warehouse strength -> Provide fast queries, governance, and reliability.
* Databricks Lakehouse merges these into one unified platform using **Delta Lake**.

* Data Lakehouse = Data Lake->(structure, semi-structure and unstructure data) + Data Warehouse(faster Queries, goverance, reliability).

##### Problems it solve :-
1. Data Silos :
  * Traditionally, companies needed a Data Lake for raw data + a Data Warehouse for analytics.
    * That meant two systems, constant ETL pipelines, and duplicate data.
    * Data Lake(for data storage) ------ Transform -----> Data Warehouse(Used for analytical purpose)
  * **LAKEHOUSE = One platform ,no silos.**

2. Reliability issues in Data Lakes :
  * Normal data lakes (S3, ADLS, GCS) don't support :-
    1. Acid transactions
    2. Schema enforcement
    3. Version control
    4. Very slow for analytics purpose.

  * This caused corrupted tables and inconsistent data.
  * Lakehouse(via Delta Lake) adds ACID transactions + schema validation + Time travel.

3. Scalability & Cost :-
  * Data Warehouse are fast but very expensive and limited to structured data.
  * Data Lakes are cheap & scalable but slow and unreliable.
  * Lakehouse keeps data in cheap cloud storage(like a lake) but allows fast SQL queries (like a waehouse).

##### How it works Internally
* Data Storage :-
  * Data is stored in cloud object storage(S3, ADLS, GCS).
  * Stored in open formats(Parquet, JSON, Avro, ORC).

* Delta Lake Layer(the magic) :-
  * Adds a transaction log (_delta_log folder) on top of files.
  * Tracks every write/update/delete.
  * Ensures ACID transactions(no partials writes).
  * Provides schema enforcement + schema evolution.
  * Enables time travel (query old version of data).

* Query & Compute Layer :-
  * Use Spark, SQL, Python, R, Scala, or Java to process data.
  * Optimized with Photon Engine(Faster SQL queries).
  * Supports batch, streaming, ML and BI workloads.
  * Governance & Security.
  * Unity Catalog manages data access, metadata, and lineage.
  * Fine-grained security across tables, columns and files.


#### Advantages of Databricks Lakehouse :-
  * Unified Platform - No need for both a lake + warehouse, everything is in one place.
  * Open Data Format - Data stays in open formats(Parquet/Delta), avoiding vendor lock-in.
  * ACID Transactions - Reliable data like a database.
  * Time travel - Query historical data versions.
  * Performance - Indexing, caching and Photon engine make queries super fast.
  * Cost-Efficient - Store on cheap cloud storage but Query like a warehouse.
  * Supports All Workloads - BI dashboards, ML training, real-time streaming, batch ETL.
  * Scalable & Elastic - Auto-Scales with cloud compute.


#### What is Unity Catalog ?
  * Unity Catalog is Databrick's governance and data management layer for all data and AI assets.
  * Think of it as a centralized system for organizing, securing, auditing and discovering data across all workspaces and cloud environments(Azure, aws, GCP).
  * It manages :
    * Tables(structured data, Delta tables)
    * Files(unstructured/semi-structured in data lake)
    * Machine Learning models
    * Dashboards
* In short: One catalog to govern everything(instead of every workspace managing its own metadat & permissions separately).
